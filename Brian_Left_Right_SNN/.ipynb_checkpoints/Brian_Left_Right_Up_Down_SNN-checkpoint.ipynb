```python
from brian2 import *
```

```python
def visualise_connectivity(S):
    Ns = len(S.source)
    Nt = len(S.target)
    figure(figsize=(10, 4))
    subplot(121)
    plot(zeros(Ns), arange(Ns), 'ok', ms=10)
    plot(ones(Nt), arange(Nt), 'ok', ms=10)
    for i, j in zip(S.i, S.j):
        plot([0, 1], [i, j], '-k')
    xticks([0, 1], ['Source', 'Target'])
    ylabel('Neuron index')
    xlim(-0.1, 1.1)
    ylim(-1, max(Ns, Nt))
    subplot(122)
    plot(S.i, S.j, 'ok')
    xlim(-1, Ns)
    ylim(-1, Nt)
    xlabel('Source neuron index')
    ylabel('Target neuron index')
```

```python
####################################################################################################
# Simulation parameters
dt = 1e-4                    # Timestep
t = np.arange(0,200000)*dt   # Simulation time array t
num_receptors = 50           # Number of receptor neurons
speed = 80                   # Speed of bright spot, number of receptor neurons passed per second
####################################################################################################

plot_t = []                  # Time of spikes
plot_n = []                  # Receptor neuron id's
currtime = 0                 # Time when a spike is fired
neuronid = 0                 # Neuron that fires a spike
stepdir = 1                  # Is the bright spot moving towards the right (1) or left (-1)
timestep = 1.0/(speed-1)     # Time between spikes generated by nearby receptor neurons

# Make placeholders for spike time arrays to be used for SNN simulation
spikes = []
for k in range(num_receptors):
    spikes.append([])

# Generate and plot spikes from receptor neurons
while currtime < t[-1]:
    plot_t.append(currtime)
    plot_n.append(neuronid)
    spikes[neuronid].append(currtime)
    if neuronid == 0 and stepdir == -1:
        stepdir = 1
    elif neuronid == (num_receptors-1) and stepdir == 1:
        stepdir = -1
    neuronid += stepdir
    currtime += timestep
```

```python
start_scope() 
spikeArr = plot_n
timeArr = plot_t*second

inputSize = num_receptors

tau = 20*ms
tau_syn = 2*ms
I_weight = 3*nA
threshold='v > -55*mV'
v_reset='v = -80*mV'
v_rest = -70*mV
R = 100*Mohm
equ = '''
dv/dt = -(v - v_rest)/tau + R*I_syn/tau : volt
dI_syn/dt = -I_syn/tau_syn : ampere
'''

generator = SpikeGeneratorGroup(inputSize, spikeArr, timeArr)
generatorMonitor = SpikeMonitor(generator)

Inputs = NeuronGroup(inputSize, equ, threshold=threshold, reset=v_reset, method=exact)
Inputs.v = -70*mV

inhib1 = NeuronGroup(inputSize-1, equ, threshold=threshold, reset=v_reset, method=exact)
inhib1.v = -70*mV

relay1 = NeuronGroup(inputSize-1, equ, threshold=threshold, reset=v_reset, method=exact)
relay1.v = -70*mV

inhib2 = NeuronGroup(inputSize-1, equ, threshold=threshold, reset=v_reset, method=exact)
inhib1.v = -70*mV

relay2 = NeuronGroup(inputSize-1, equ, threshold=threshold, reset=v_reset, method=exact)
relay2.v = -70*mV

output = NeuronGroup(2, equ, threshold=threshold, reset=v_reset, method=exact)
output.v = -70*mV

GenToIn = Synapses(generator, Inputs, on_pre='I_syn += I_weight')
GenToIn.connect(condition='j == i')
#visualise_connectivity(GenToIn)

##################################################################################################

InToInh1 = Synapses(Inputs, inhib1, on_pre='I_syn += I_weight')
InToInh1.connect(condition='j+1 == i')
#visualise_connectivity(InToInh1)

InToRel1 = Synapses(Inputs, relay1, on_pre='I_syn += I_weight')
InToRel1.connect(condition='j == i')
#visualise_connectivity(InToRel1)

Inh1ToRel1 = Synapses(inhib1, relay1, on_pre='I_syn -= I_weight')
Inh1ToRel1.connect(condition='j == i')
#visualise_connectivity(Inh1ToRel1)

Rel1Toout = Synapses(relay1, output, on_pre='I_syn += I_weight/3')
Rel1Toout.connect(condition='j == 0')
#visualise_connectivity(Rel1Toout)

##################################################################################################

InToInh2 = Synapses(Inputs, inhib2, on_pre='I_syn += I_weight')
InToInh2.connect(condition='j == i')
#visualise_connectivity(InToInh2)

InToRel2 = Synapses(Inputs, relay2, on_pre='I_syn += I_weight')
InToRel2.connect(condition='j+1 == i')
#visualise_connectivity(InToRel2)

Inh2ToRel2 = Synapses(inhib2, relay2, on_pre='I_syn -= I_weight')
Inh2ToRel2.connect(condition='j == i')
#visualise_connectivity(Inh2ToRel2)

Rel2Toout = Synapses(relay2, output, on_pre='I_syn += I_weight/3')
Rel2Toout.connect(condition='j == 1')
#visualise_connectivity(Rel2Toout)

##################################################################################################

outputMonitor = SpikeMonitor(output)
outputState = StateMonitor(output, 'v', record=True)

relayMonitor = SpikeMonitor(relay1)
relayState = StateMonitor(relay1, 'v', record=True)

run(1000*ms)
```

```python
plot(outputState.t/ms, outputState.v[0], label='output1')
plot(outputState.t/ms, outputState.v[1], label='output2')
```

```python
plot(relayState.t/ms, relayState.v[0], label='Relay1')
plot(relayState.t/ms, relayState.v[1], label='Relay2')
plot(relayState.t/ms, relayState.v[2], label='Relay3')
plot(relayState.t/ms, relayState.v[3], label='Relay4')
xlabel('Time (ms)')
ylabel('v')
legend()
```
